```python
# Bước 1: Đảm bảo cột Price đang là chuỗi (string)
df['Price'] = df['Price'].astype(str)

# Điều kiện: chứa ký tự chữ hoặc ký tự đặc biệt, không phải toàn số/chấm

invalid_prices = df[~df['Price'].str.match(r'^\d+(\.\d+)?$', na=False)]

# Bước 3: In ra những giá trị bất thường
print("📌 Các giá trị Price không thể chuyển sang số:")
print(invalid_prices['Price'].unique())

# Nếu muốn xem cả dòng
print("📌 Các dòng bị lỗi khi chuyển Price:")
print(invalid_prices[['Price']].head())
print(f"Số lượng giá trị không hợp lệ trong Price: {len(invalid_prices)}")
```

```python
# Bước 1: Đảm bảo cột Price là chuỗi
df['Price'] = df['Price'].astype(str)

# Bước 2: Tạo mask để tìm các dòng không thể chuyển sang số
invalid_mask = ~df['Price'].str.match(r'^\d+(\.\d+)?$', na=False)

# Bước 3: Giữ lại dòng hợp lệ, nhưng giữ nguyên cột gốc
df_cleaned = df[~invalid_mask].copy()

# Bước 4: Tạo cột mới: Price_vnd = Price đã chuyển sang số * 26,000
df_cleaned['Price_vnd'] = pd.to_numeric(df_cleaned['Price']) * 26000
```

```python
df_cleaned.isnull().sum()
```

```python
df_cleaned = df_cleaned.drop_duplicates()
```

```python
def check_invalid(col, reg=r'^[0-9]+$', replace_reg=r'[0-9]'):
    # Ép kiểu chuỗi để tránh lỗi .str
    col = col.astype(str)

    invalid_row = col[~col.str.match(reg)]
    invalid_count = invalid_row.count()

    invalid = col.str.replace(replace_reg, '', regex=True).unique()

    return invalid, invalid_count
```

```python
df_cleaned.dropna(subset=['Reg_year'], inplace=True)
```

```python
df_cleaned.isnull().sum()
```

```python
cols = ['Color','Bodytype', 'Gearbox', 'Fuel_type']
df = df_cleaned.dropna(subset=cols)
df.isnull().sum()
```

```python
df = df[~df['Genmodel'].str.contains(r'\d{2,4}[-/]\d{1,2}[-/]\d{1,4}', na=False)].copy()
```

```python
# # Bước 2: Chọn các cột cần kiểm tra outlier
# cols_to_check = ['Runned_Miles', 'Engin_size']

# # Bước 3: Vẽ biểu đồ boxplot
# plt.figure(figsize=(12, 5))

# for i, col in enumerate(cols_to_check, 1):
#     plt.subplot(1, 2, i)
#     sns.boxplot(y=df[col], color='skyblue')
#     plt.title(f'Boxplot: {col}')

# plt.tight_layout()
# plt.show()
```

```python
df['Engin_size'] = df['Engin_size'].astype(str)

invalid_runned = df[~df['Engin_size'].str.match(r'^\d+(\.\d+)?$', na=False)]

print("❌ Các giá trị không hợp lệ trong Engin_size:")
print(invalid_runned['Engin_size'].unique())
print(f"Số lượng giá trị không hợp lệ trong Price: {len(invalid_runned)}")
```

```python
df['Engin_size'] = df['Engin_size'].str.replace('L', '', regex=False)
df['Engin_size'] = pd.to_numeric(df['Engin_size'], errors='coerce')  # Nếu có lỗi sẽ ra NaN
```

```python
df['Engin_size'] = df.groupby('Genmodel')['Engin_size'].transform(lambda x: x.fillna(x.median()))
df.isnull().sum()
```

```python
df['Runned_Miles'] = df['Runned_Miles'].astype(str)

invalid_runned = df[~df['Runned_Miles'].str.match(r'^\d+(\.\d+)?$', na=False)]

print("❌ Các giá trị không hợp lệ trong Engin_size:")
print(invalid_runned['Runned_Miles'].unique())
print(f"Số lượng giá trị không hợp lệ trong Price: {len(invalid_runned)}")
```

```python
df['Runned_Miles'] = pd.to_numeric(df['Runned_Miles'], errors='coerce')
df['Runned_Miles'].fillna(df['Runned_Miles'].median(), inplace=True)
df.isnull().sum()
```

```python
for col in ['Seat_num', 'Door_num']:
    mode_val = df[col].mode()[0]         # lấy mode đầu tiên
    df[col].fillna(mode_val, inplace=True)
```

```python
df = df[~df['Engin_size'].isnull()].copy()
df.isnull().sum()
```

```python
df['Price'] = pd.to_numeric(df['Price'], errors='coerce')
```

#Unusual case

```python
# Seat vs Door không hợp lý
print("🚨 Seat_num = 2 mà Door_num >= 4:")
print(df[(df['Seat_num'] == 2) & (df['Door_num'] >= 4)])


```

```python
invalid_seat_door = (df['Seat_num'] == 2) & (df['Door_num'] >= 4)

df = df[~invalid_seat_door].copy()
```

```python
# Xe đăng ký sau khi rao bán - vì là xe cũ nên không hợp lý
print("🚨 Năm đăng ký lớn hơn năm rao bán:")
print(df[df['Reg_year'] > df['Adv_year']])


```

```python
invalid_reg_adv = df['Reg_year'] > df['Adv_year']

df = df[~invalid_reg_adv].copy()
```
